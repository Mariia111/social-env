```python
df = unemployment[['Country','2010','2011','2012','2013','2014','2015','2016','2017']]
df
```

```python
belarus = df[df['Country'] == 'Belgium']
value = belarus.values.tolist()[0][1:]
column = belarus.columns.tolist()[1:]
column
```

```python
value = [float(i) for i in value]
value
```

```python
plt.plot(column, value, label = 'Уровень безработици')
title_dict = {'fontsize':20, 'fontweight':'bold', 'color':'red', 'family': 'calibri'}
plt.title('Безработица', fontdict=title_dict)

fig_size = plt.gcf().get_size_inches()
sizefactor = 3 
plt.gcf().set_size_inches(sizefactor * fig_size)
plt.xlabel('Года')
plt.ylabel('Безработица')
plt.grid()
plt.show()
```
